# Face Detection


## Steps to follow


#### 1. Get A Crap-Load of Faces ( to train your model perfectly )
#### 2. Convert into grayscale ( opencv )
#### 3. Train the Algorithm to detect faces

## Importing the  libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Loading the data 

```python
# Load some pre-trained data on face frontals from opencv ( haar cascade algorithm )
trained_face_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml');
```

## Testing the face detection 

```python
# Loading an image to check if its working fine or not
img = cv2.imread('friends.png')
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
# BGR to RGB
grid_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
# BGR to RGB
grayScale = cv2.cvtColor(grid_RGB, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(20,8))
plt.imshow(grayScale)
```

```python
from random import randrange

# Detect Faces
face_coordinates = trained_face_data.detectMultiScale(grayScale)

# Draw rectangles around the faces
for (x ,y ,w ,h) in face_coordinates:
    cv2.rectangle(img,(x,y),(x+w , y+h),(0,100,randrange(220,255)),3)
```

### Displaying the image using opencv

```python
# Display the faces with detection
cv2.imshow("Automatic Face Detector",img)
cv2.namedWindow("Automatic Face Detector",cv2.WINDOW_NORMAL)
cv2.resizeWindow("Automatic Face Detector", 1000,700)
cv2.waitKey()
```
